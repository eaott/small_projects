In this notebook, I want to explore some scenarios for student loans for a message I want to give to the youth group on money. I wanted to explore some options about how long it takes to pay off student loans based on average rates, etc.

```python
import numpy as np
import matplotlib.pyplot as plt
def example(initial_principal, rate, payment):
    principal = initial_principal
    paid = 0

    balance = [principal]
    payments = [0]
    while principal > 0:
        monthly_interest = principal * rate / 365.0 * 30
        delta = payment - monthly_interest
        if delta > 0:
            principal -= delta
        balance.append(principal)
        payments.append(payment)

    fig, ax = plt.subplots(1, figsize=(7,7))
    age = 22 + np.array(range(len(balance))) / 12.0
    ax.plot(age, balance, label="remaining balance")
    ax.plot(age, np.cumsum(payments), label="total paid")
    ax.plot([age[0], age[-1]], [initial_principal, initial_principal], "k", label="initial loan")
    ax.legend(loc="best")
    ax.set_xlabel("age")
    ax.set_ylabel("amount ($)")
    ax.set_title("\$%dk %0.2f%% loan with \$%d monthly payments" % (initial_principal / 1000, rate * 100, payment))

    ax.text(0.65, 0.6, "Total cost: \$%s" % "{:,}".format(np.sum(payments)), transform=ax.transAxes)
    ax.text(0.65, 0.55, "Total time: %d years" % np.round(len(payments) / 12.0), transform=ax.transAxes)
    ax.text(0.65, 0.5, "Total interest: \$%0.0f" % (np.sum(payments) - initial_principal), transform=ax.transAxes)
    plt.show()
```

```python
example(30000, 0.04, 100)
```

```python
example(30000, 0.0505, 150)
```

```python
example(10000, 0.0505, 150)
```

```python
example(60000, 0.08, 400)
```
